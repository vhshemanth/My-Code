```python
X=df.drop(['pts'],axis=1)
y=df['pts']
```

```python
q = X['fg'].quantile(0.99)
data=X.copy()
# we are removing the top 2% data 
data_cleaned = data[data['fg']<q]
q = data_cleaned['x3p'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['x3p']<q]
q = data_cleaned['x2p'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['x2p']<q]

q = data_cleaned['ft'].quantile(0.99)
# we are removing the top 1% data from the BMI column
data_cleaned  = data_cleaned[data_cleaned['ft']<q]
q = data_cleaned['fta'].quantile(0.99)
# we are removing the top 1% data
data_cleaned  = data_cleaned[data_cleaned['fta']<q]
q = data_cleaned['orb'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['orb']<q]
q = data_cleaned['trb'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['trb']<q]
q = data_cleaned['ast'].quantile(0.99)
data_cleaned  = data_cleaned[data_cleaned['ast']<q]
q = data_cleaned['stl'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['stl']<q]
q = data_cleaned['blk'].quantile(0.99)
# we are removing the top 1% data 
data_cleaned  = data_cleaned[data_cleaned['blk']<q]
q = data_cleaned['tov'].quantile(0.99)
# we are removing the top 1% data
data_cleaned  = data_cleaned[data_cleaned['tov']<q]
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=data_cleaned)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=11)
```

```python
X.isnull().sum()
```

```python
X['x2p.']=X['x2p.'].fillna(X['x2p.'].mean())
X['efg.']=X['efg.'].fillna(X['efg.'].mean())
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=pred)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```
